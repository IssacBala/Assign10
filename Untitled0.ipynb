```
import numpy as np
import pandas as pd
import datetime as dt
from keras.layers.core import Dense, Activation, Dropout, Flatten
from keras.layers.recurrent import LSTM
from keras.models import Sequential
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import math
import io
from google.colab import files
uploaded = files.upload()
```

```
# url = 'https://github.com/IssacBala/Assign10/blob/master/GoogleStocks.csv'
# df1 = pd.read_csv(url)
# df1.values

df = pd.read_csv(io.StringIO(uploaded['GoogleStocks.csv'].decode('utf-8')),thousands=',')
```

```
df['date'] = pd.to_datetime(df['date'])
df['volume'] = pd.to_numeric(df['volume'])
df['high'] = pd.to_numeric(df['high'])
df['low'] = pd.to_numeric(df['low'])
df['avg'] = (df['high']+df['low'])/2
df.set_index('date',inplace=True)
# df.info()
df.sort_index(inplace=True)
```

```
plt.plot(df['avg'].values)
plt.show()
```

```
scalerAvg = MinMaxScaler(feature_range=(0,1))
acDf = scalerAvg.fit_transform(df[['avg','volume','open']].values)
```

```
def seriesDS(dfts,sLength):
  X,Y = [],[]
  for i in range(dfts.shape[0]-sLength+1):
    X.append(dfts[i:(i+sLength),:2])
    Y.append(dfts[(i+sLength)-1,2])
  return np.array(X), np.array(Y)

trSize = int(len(df)* 0.7)
teSize = len(df) - trSize
train, test = acDf[0:trSize, :], acDf[trSize:len(acDf), :]
print(trSize,teSize,len(df),train.shape,test.shape)
```

```
a = np.arange(21).reshape(7,3)
print(len(a),a)
seriesDS(a,3)
```

```
def plotthePred(actual,step,trainYb,testYb ):
  print(step,len(trainYb)+step,len(trainYb)+(step*2)+1,len(actual)-1,len(testYb),len(actual)-1-len(testYb))
  plot1 = np.empty_like(actual)
  plot1[:] = np.nan
  plot1[step:len(trainYb)+step,2] = trainYb[:,2]
  plot2 = np.empty_like(actual)
  plot2[:] = np.nan
  plot2[len(actual)-1-len(testYb):len(actual)-1,2] = testYb[:,2]
  inv = scalerAvg.inverse_transform(actual)
  plt.plot(inv[:,2])
  plt.plot(plot1[:,2])
  plt.plot(plot2[:,2])
  plt.show()
```

```
hidden=np.array([2,3])
cells=np.array([30,50,80])
steps=np.array([20,50,75])
model, predictTr, predictTe,trainScore,testScore,trainPred,testPred = {}, {}, {}, {}, {}, {}, {}
count =0
trainAct = scalerAvg.inverse_transform(train)
testAct = scalerAvg.inverse_transform(test)
for s in steps:
  print(train.shape,test.shape)
  trainX, trainY = seriesDS(train,s)
  testX,testY = seriesDS(test,s)
  testY = testY.reshape(teSize-s+1,1)
  trainY = trainY.reshape(trSize-s+1,1)
  print(trainX.shape,trainY.shape)
#   testX = testX.reshape(teSize-s,s,2)
#   trainX = trainX.reshape(trSize-s,s,2)
  for c in cells:
    for h in hidden:
      print('Model = ',h,' hidden layers. ',c,' cells, ',s,' time steps')
      count = count +1
      model[count] = Sequential()
      model[count].add(LSTM(units = c, return_sequences = True, input_shape=(s, 2)))
      model[count].add(Dropout(0.2))
      for i in np.arange(h-1):
        model[count].add(LSTM(units = c, return_sequences = True))
        model[count].add(Dropout(0.2))
      model[count].add(Flatten())
      model[count].add(Dense(units = 1))
      model[count].compile(optimizer='adam', loss='mse')
      model[count].fit(trainX, trainY, epochs = 100, batch_size = 32)
      predictTr[count] = model[count].predict(trainX)
      predictTe[count] = model[count].predict(testX)
      #unscale predictions
      #print(train[s:,:2].shape,predictTr[count].shape)
      trainPred[count] = scalerAvg.inverse_transform(np.concatenate((train[s-1:,:2],predictTr[count]),axis=1))
      testPred[count] = scalerAvg.inverse_transform(np.concatenate((test[s-1:,:2],predictTe[count]),axis=1))

      #print(trainPred[count].shape,trainAct[:,2].shape, testPred[count][:,2].shape)
      trainScore[count] = math.sqrt(mean_squared_error(trainPred[count][:,2], trainAct[s-1:, 2]))
      testScore[count] = math.sqrt(mean_squared_error(testPred[count][:,2], testAct[s-1:, 2]))
      print('Train score: %.2f rmse', trainScore[count])
      print('Test score: %.2f rmse', testScore[count])
      #print(trainPred[count].shape,testPred[count].shape)
      plotthePred(acDf,s,trainPred[count],testPred[count])
```

```
count =0
axisList = []
trainScoreAr = np.zeros((2,3,3))
testScoreAr = np.zeros((2,3,3))
for s in steps:
  for c in cells:
    for h in hidden:
      count = count+1
      trainScoreAr = trainScore[count]
      axisList.append('('+str(h)+','+str(c)+','+str(s)+')')
      
fig1, ax = plt.subplots(figsize=(20,6), dpi= 80, facecolor='w', edgecolor='k')
trainScoreAr = np.array(list(trainScore.values()))
testScoreAr = np.array(list(testScore.values()))
index = np.arange(18)
bar_width = 0.12
opacity = 0.8
rects1 = plt.bar(index, trainScoreAr, bar_width,alpha=opacity,color='b',label='Train Score')
rects2 = plt.bar(index + bar_width, testScoreAr, bar_width,alpha=opacity,color='g',label='Test Score')
plt.xlabel('Type')
plt.ylabel('Scores')
plt.title('Scores by Type')
plt.xticks(index + bar_width, axisList)
plt.legend()
 
plt.tight_layout()
plt.show()
```
